```python
import numpy as np
import pandas as pd 

import matplotlib
import matplotlib.pyplot as plt 
```

```python
fdir = "/mnt/raid-cita/dtolgay/doga_duo/runs/zoom_in/z0.0/m12i_res7100_md/voronoi_1e6/radmc3d_in.txt"

radmc3d = pd.DataFrame(np.loadtxt(fdir), columns = ["x", "y", "z", "vx", "vy", "vz", "nh2", "Tdust", "Tgas"])
```

```python
R_max = 20e3 # pc

print("Plotting galaxy")
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # Adjust figsize as needed

# First subplot: x-y axis plot
axs[0].hist2d(
    x=radmc3d["x"],
    y=radmc3d["y"],
    bins=500,
    norm=matplotlib.colors.LogNorm(),
    range=[[-R_max, R_max], [-R_max, R_max]]
)
axs[0].set_title('x-y axis plot')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Second subplot: z-y axis plot
axs[1].hist2d(
    x=radmc3d["y"],  # Use the 'y' column for the x-axis
    y=radmc3d["z"],  # Keep the 'z' column for the y-axis
    bins=500,
    norm=matplotlib.colors.LogNorm(),
    range=[[-R_max, R_max], [-R_max, R_max]]
)
axs[1].set_title('z-y axis plot')
axs[1].set_xlabel('y')
axs[1].set_ylabel('z')
```
